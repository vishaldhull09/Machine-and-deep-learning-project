<h1>Removing Punctuation and converting it into lower case</h1>

```python
combined = [train,test]

for data in combined:
    data1 = data.iloc[:,2:27]
    data1.replace('[^A-Za-z]'," ",regex=True,inplace=True)
    data.loc[:,2:27] = data1.loc[:,:]
    
train.head()    
```

```python
for data in combined:
    data1 = data.iloc[:,2:27]
    for col in data1.columns:
        data1[col] = data1[col].str.lower()
    data.loc[:,2:27] = data1.loc[:,:]
```

<h1>practice of CountVectorize</h1>

```python
 
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
     'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
     'Is this the first document?']

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())
#['and', 'document', 'first', 'is', 'one', 'second', 'the', 'third', 'this']
print(X.toarray())  # doctest: +NORMALIZE_WHITESPACE
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
     'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
     'Is this the first document?']

vectorizer2 = CountVectorizer(ngram_range=(2,2))
X = vectorizer2.fit_transform(corpus)
print(vectorizer2.get_feature_names())
#['and', 'document', 'first', 'is', 'one', 'second', 'the', 'third', 'this']
print(X.toarray())  # doctest: +NORMALIZE_WHITESPACE
```

<h1>Classification using RandomForestClassifier</h1>

```python
len(headline1)
```

```python
count_vector = cv(ngram_range=(2,2))
traindataset = count_vector.fit_transform(headline1)

clf = RandomForestClassifier(n_estimators=200,criterion='entropy')
clf.fit(traindataset,train['Label'])
```

```python
test_data = count_vector.transform(headline2)

prediction = clf.predict(test_data)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

matrix = confusion_matrix(test['Label'],prediction)
print(matrix)

score = accuracy_score(test['Label'],prediction)
print(score)

report = classification_report(test['Label'],prediction)
print(report)
```
